```python
# Create a series of three different colours
colours = pd.Series(["purple","pink","blue"])
```

```python
# Create a series of three different car types and view it
car_types =pd.Series(["BMW","Tesla","porsche"])
car_types
```

```python
# Combine the Series of cars and colours into a DataFrame
combined = pd.DataFrame({"car_type": car_types,"colour":colours})
combined
```
