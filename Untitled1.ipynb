```python
What nonlinear means

    Uses kernel functions
    Transforms data into a higher-dimensional space
    Each pixel is compared with many reference points (training samples)

How NELM behaves

    Similar pixels may get very different outputs
    Can separate complex class boundaries
    More flexible, more powerful

Strengths
    
    Better spectral-only accuracy
    Handles complex class overlap
    Stronger decision boundaries

Weaknesses

    Distorts feature relationships
    Sensitive to noise and small differences
    Can break neighborhood consistency

When NELM is good

    When only spectral classification is used
    When spatial consistency is NOT explicitly modeled
```
