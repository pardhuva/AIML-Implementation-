 # Anatomy of a matplotlib figure

### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt 

# 1. prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot 
fig, ax = plt.subplots(figsize=(10,10))  #(width,height)

# 3. plot data 
ax.plot(x,y)

# 4. Customize plot 
ax.set(title="Simple plot",
        xlabel="x-axis",
        ylabel="y-axis")
# 5. save & show ( you save teh whole figure)
fig.savefig("nh.jpg")
plt.show()
```

```python
# lets try some example 
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [10,20,30,40]

fig,ax= plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="some random example",
       xlabel="x-axis",
       ylabel="y-axis"
       )
```

```python
# make a plot from dictionary 
nut_butter_prices = { "Almond butter":10,
                      "Peanut butter": 8,
                      "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())   #(x,y)
ax.set(title ="Pardhuva nut butter store",
       ylabel="Price",
       xlabel="butter type");
```

### Two options for subplots

```python
# subplot option1 
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# plot to each different axis 
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000))
```

```python
# subplots option2
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
#remove those last two zeroes
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]= car_sales["Price"].cumsum()
car_sales
```

```python
# so price is still in string type
type(car_sales["Price"][0])
```

```python
# so now we have to convert to integer
car_sales["Total Sales"]= car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# lets plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# now scatter plot 
car_sales["Price"] = car_sales["Price"].astype(int)    # before plotting make sure it is integer
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
# plotting from pandas dataframes6
```

```python
# OO method from scratch
fi, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart disease and cholestrol levels",
       xlabel='Age',
       ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# to add a horizontal line indicatig a average
# how to add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--')    # we have to give y axis here mean indicate what line we want
```

```python
over_50.target.values
```

```python
# plotting from pandas dataframes 7
```

```python
# lets add subplot to the above plot
# subplot of chol,age,thalach
fig, (ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10
                                                     ))    # single figure with two axes 

# Adding data to the axes
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# add data to ax1 
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
#customize ax1
ax1.set(xlabel='Age',ylabel='Thalach',title="Heart disease and Max Heart rate")

# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# add a meanline 
ax1.axhline(y=over_50["thalach"].mean(),linestyle='--')

# Add a title to the figure 
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
# for different styles 
# cmap = "winter"   # this changes the colour scheme
# OO method from scratch
import matplotlib.pyplot as plt
fi, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="plasma")

# customize the plot
ax.set(title="Heart disease and cholestrol levels",
       xlabel='Age',
       ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# to add a horizontal line indicatig a average
# how to add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--')    # we have to give y axis here mean indicate what line we want
```

```python
# customizing the x and y axis limitations
# lets add subplot to the above plot
# subplot of chol,age,thalach
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10
                                                     ))    # single figure with two axes 

# Adding data to the axes
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")

#customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        ylabel="Cholesterol")

#change the x-axis limitations
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# add data to ax1 
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                     cmap="winter")
#customize ax1
ax1.set(xlabel='Age',ylabel='Thalach',title="Heart disease and Max Heart rate")

# change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# add a meanline 
ax1.axhline(y=over_50["thalach"].mean(),linestyle='--')

# Add a title to the figure 
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
