# Dot product vs Element wise

```python
mat1
```

```python
mat2
```

```python
mat1.shape, mat2.shape
```

```python
# elementwise multiplication (hadamard product)
mat1 * mat2
```

```python
# dot product
np.dot(mat1,mat2)
```

```python
# element wise is like normal multiplication like 
# suppose [1,2,3][4,5,6] then [1*4,2*5,3*6] in this way for each multiply with their corresponding  row
# for dot prouduct it does matrix multiplication 
# we know for matrix multiplication the inner dimensions should match 
# (a,b) (b,c) then on mulplication it becomes (a,c)
```

```python
# Transpose mat1
mat1.T
```

```python
mat2.shape, mat1.T.shape
```

```python
# so if we do transpose then the inner dimensions mamtch 
mat3 = np.dot(mat1, mat2.T)
```

# Dot product example (nut butter sales)

```python
np.random.seed(0)

# Number of jars sold 
sales_amounts = np.random.randint(20,size=(5,3))
sales_amounts
```

```python
# create a weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                            index=["Mon","Tues","Wed","Thurs","Fri"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
weekly_sales
```

```python
# create prices array 
prices = np.array([10,8,12])
prices
```

```python
# create butter prices DataFrame 
butter_prices = pd.DataFrame(prices.reshape(1,3),
                             index=["Price"],
                             columns=["Almond butter", "Peanut butter", "Cashew butter"]
                            )
butter_prices
```

```python
prices, prices.shape
```

```python
sales_amounts, sales_amounts.shape
```

```python
total_sales = prices.dot(sales_amounts)
```

```python
# shapes arent aligned, lets transpose 
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
sales_amounts.T.shape
```

```python
prices.shape
```

```python
# this is easier version of above all 
np.dot(sales_amounts,prices)
```

```python
# A NumPy 1-D array (n,) is treated as a row or column vector based on the operation and 
 #its position in that operation.
# so if (3,) then it can be treated as (3,1) or (1,3) based on the operation 

# if (3,) is on the left of the dot product then it is treated as row vector 
# ex: (3,) . (3,5)   =>     (1,3) . (3,5)

#  if (3,) is on the right of the dot product then it is treated as column vector 
# ex: (5,3) . (3,)   =>     (5,3) . (3,1)
```

```python
weekly_sales
```

```python
weekly_sales["Total cost ($)"] = total_sales
weekly_sales
```

```python
weekly_sales.drop("Total cost ($)", axis=1, inplace=True)
weekly_sales
```

```python
butter_prices
```

```python
weekly_sales["Total cost ($)"] = total_sales
weekly_sales
```
