```python
# inplace parameter 
# sometimes data can have missing data (some data is not filled in some columns 
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
mean = car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing
```

```python
# This is because we have to inplace modifying 
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(mean,inplace=True)
```

```python
# you can remove the rows with NaN , this might remove inplace 
car_sales_missing.dropna(inplace=True)
```

# Creating data from previous Existing data 

```python
# Column from series 
seats_column = pd.Series([5,5,5,5,5])

# new column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
# fill those seats with 
car_sales["Seats"].fillna(5,inplace=True)
```

```python
car_sales
```

```python
# column from python list 
fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.5,6.7,8.9,7.9,1.3]
car_sales["Fuel per 10 KM"] = fuel_economy
car_sales
```

```python
# creating column from other columns 
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100*car_sales["Fuel per 10 KM"]
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100*car_sales["Fuel per 10 KM"]
car_sales
```

```python
# creating column from single value 
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road test"] = True
car_sales
```
